STEP 03 Processing DataSet

```python
#df.dropna() drop all null values
```

```python
mean_e = df.YearsExperience.mean()
mean_s = df.Salary.mean()

print(mean_e," Mean of Experience")
print(mean_s," Mean of Salary")
```

```python
import math

df.YearsExperience = df.YearsExperience.fillna(math.floor((mean_e)))
df.Salary = df.Salary.fillna(math.floor((mean_s)))

df
```

```python
import seaborn as sb

def get_heat_map(df):
  sb.heatmap(df.isnull() , cmap='viridis')
  return plt.show()

get_heat_map(df)
```

#Step 04 Loading DATA

```python
#Method 01
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1:].values

X
```

```python
#Method 02
import numpy as np

X = np.array(df[['YearsExperience']])
Y = np.array(df[['Salary']])

X
```

```python
#Step 05 Splitting Data - Train Set & Teat Set
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)

X_train
```

#Step 06 Tain Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
x_pred = model.predict(Y_test)
y_pred
```

#Step 07 Performance Evaluation

```python
model.score(X_test , Y_test)*100
```

```python
model.score(X_train , Y_train)*100
```

```python
model.score(X , Y)*100
```
